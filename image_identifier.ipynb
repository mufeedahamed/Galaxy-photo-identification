```python
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import cv2

```

```python
# mount
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load the saved model
model = tf.keras.models.load_model("/content/drive/My Drive/Colab Notebooks/6/galaxy_classification_model.h5")

# Load the test images
test_images = []
for i in range(10): # load the first 10 test images for example purposes
    img_path = f'/content/drive/My Drive/Colab Notebooks/6/Valid/Real/img_{i}.png' # replace with the actual path to the test image
    img = cv2.imread(img_path)
    test_images.append(img)

# Convert the test images to a numpy array and normalize
test_images = np.array(test_images) / 255.0

# Use the loaded model to predict the labels for the test images
preds = model.predict(test_images)

# Print the predicted labels
print('Predictions:')
for i in range(10):
    h = f'Test image {i}: Real' if preds[i] > 0.5 else f'Test image {i}: Fake'
    print (h)

```

```python

# Plot the images with their predicted labels
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(12, 6))
for i, ax in enumerate(axes.flat):
    # Plot the image
    ax.imshow(test_images[i])
    # Set the title with the predicted and true labels
    ax.set_title(f'Test image {i}: Real' if preds[i] > 0.5 else f'Test image {i}: Fake')
    ax.axis('off')
plt.show()
```
